```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('bankloan.csv')
df
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df[['ID','Age','Experience','Income','ZIP.Code','Family','CCAvg','Education','Mortgage','Personal.Loan','Securities.Account','CD.Account','Online']]
X
```

```python
y=df['CreditCard']
y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33 ,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
predictions =model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import accuracy_score
```

```python
score=accuracy_score(y_test , predictions)
score
```

```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QPushButton, QTextEdit, QHBoxLayout, QMessageBox
from sklearn.metrics import classification_report, accuracy_score
from PyQt5.QtGui import QFont
from PyQt5.QtCore import Qt

class MetricsApp(QWidget):
    def __init__(self, y_train, predictions):
        super().__init__()

        self.y_train = y_train
        self.predictions = predictions

        self.initUI()

    def initUI(self):
        # Layout and Widgets
        self.setWindowTitle('BANK LOAD PREDICTION MODEL')
        self.setGeometry(300, 200, 600, 500)

        # Styling
        self.setStyleSheet("background-color: #f0f0f0;")

        # Text Area for Results
        self.text_edit = QTextEdit(self)
        self.text_edit.setReadOnly(True)
        self.text_edit.setFont(QFont("Courier", 10))
        self.text_edit.setStyleSheet("border: 2px solid #0078D7; padding: 10px;")

        # Buttons
        self.show_metrics_btn = QPushButton('Show Metrics', self)
        self.show_metrics_btn.setStyleSheet("background-color: #8BC34A; color: white; padding: 10px;")  # Light Green
        self.show_metrics_btn.setFont(QFont("Arial", 12, QFont.Bold))
        self.show_metrics_btn.clicked.connect(self.display_metrics)

        self.clear_btn = QPushButton('Clear', self)
        self.clear_btn.setStyleSheet("background-color: #FF9800; color: white; padding: 10px;")  # Orange
        self.clear_btn.setFont(QFont("Arial", 12, QFont.Bold))
        self.clear_btn.clicked.connect(self.clear_text)

        self.quit_btn = QPushButton('Quit', self)
        self.quit_btn.setStyleSheet("background-color: #F44336; color: white; padding: 10px;")  # Red
        self.quit_btn.setFont(QFont("Arial", 12, QFont.Bold))
        self.quit_btn.clicked.connect(self.close_app)

        self.help_btn = QPushButton('Help', self)
        self.help_btn.setStyleSheet("background-color: #3F51B5; color: white; padding: 10px;")  # Dark Blue
        self.help_btn.setFont(QFont("Arial", 12, QFont.Bold))
        self.help_btn.clicked.connect(self.show_help)

        # Layout management
        layout = QVBoxLayout()

        # Add text area
        layout.addWidget(self.text_edit)

        # Buttons at the bottom
        button_layout = QHBoxLayout()
        button_layout.addWidget(self.show_metrics_btn)
        button_layout.addWidget(self.clear_btn)
        button_layout.addWidget(self.help_btn)
        button_layout.addWidget(self.quit_btn)

        layout.addLayout(button_layout)
        self.setLayout(layout)

    def display_metrics(self):
        try:
            accuracy = accuracy_score(self.y_train, self.predictions)
            report = classification_report(self.y_train, self.predictions)

            result_text = f"Accuracy: {accuracy:.2f}\n\nClassification Report:\n{report}"
            self.text_edit.setText(result_text)
        except Exception as e:
            self.show_error(str(e))

    def clear_text(self):
        self.text_edit.clear()

    def show_help(self):
        msg = QMessageBox()
        msg.setIcon(QMessageBox.Information)
        msg.setWindowTitle("Help")
        msg.setText("This app calculates accuracy and classification report from a dataset.\n\n"
                    "Steps:\n"
                    "1. Load your data.\n"
                    "2. Press 'Show Metrics' to display results.\n"
                    "3. Use 'Clear' to clear the text area.\n"
                    "4. Press 'Quit' to exit the app.")
        msg.setStandardButtons(QMessageBox.Ok)
        msg.exec_()

    def close_app(self):
        QApplication.quit()

    def show_error(self, error_message):
        msg = QMessageBox()
        msg.setIcon(QMessageBox.Critical)
        msg.setWindowTitle("Error")
        msg.setText("An error occurred!")
        msg.setInformativeText(error_message)
        msg.setStandardButtons(QMessageBox.Ok)
        msg.exec_()

if __name__ == '__main__':
    # Dummy data for demonstration (replace with your actual y_train and predictions)
    y_train = [0, 1, 0, 1, 0, 1]
    predictions = [0, 1, 0, 0, 0, 1]

    # Create the app
    app = QApplication(sys.argv)
    metrics_app = MetricsApp(y_train, predictions)
    metrics_app.show()

    # Run the application
    sys.exit(app.exec_())
```
