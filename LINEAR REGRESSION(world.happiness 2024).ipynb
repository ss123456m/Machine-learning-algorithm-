```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv("World Happiness Report 2024.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# droping the null values
df1=df.dropna()
df1.isnull().sum()
```

```python
df1.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df1[['year','Life Ladder','Log GDP per capita','Social support','Healthy life expectancy at birth','Freedom to make life choices','Generosity','Perceptions of corruption','Positive affect']]
X
```

```python
y=df1['Negative affect']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.20,random_state=5)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
predictions =model.predict(X_test)
```

```python
errors =np.abs(predictions -y_test)
plt.scatter(predictions,y_test,c=errors ,cmap='coolwarm',s=50)  # color for errors
plt.show()
```

```python
print("mean absolute error :",mean_absolute_error(y_test,predictions))
print("Mean Squared Error :",mean_squared_error(y_test ,predictions))
```

```python
result_of_data = pd.DataFrame({
    'Actual':y_test,
    'Predicted':predictions,
    'Difference':y_test-predictions
})
```
